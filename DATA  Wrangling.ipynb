# Method of Normalization
1. Simple feature scaling
      x(new)= x(old)/x(max(max)
2. min-max method of Normalization
3. Z-score (standard score) -3 to +3
4. Log transformationb

```python
# simple feature scaling
ks4['fare'] = ks4['fare']/ks4['fare'].max()
ks4.head()
```

```python
# min -max method

ks4['fare'] = (ks4['fare']-ks4['fare'].min()) / (ks4['fare'].max()-ks4['fare'].min()) 
ks4.head()
```

```python
# Z score  -3  to +3
 ks4['fare'] = (ks4['fare']-ks4['fare'].mean()) /ks4['fare'].std()
```

```python
# log transformation
ks4['fare'] = np.log(ks4['fare'])
ks4.head()
```

## Binning
- grouping of values into smaller number of values(bins)
- convert numeric inti categories(child, young, old) or 1-16, 17-30 etc to integer
- to have better understanding of of groups
     low vs mid vs high 

```python
bins = np.linspace(min(kashti["age"]), max(kashti["age"]), 15000)
age_groups = ["child", "young", "old"]
kashti['age'] = pd.cut(kashti["age"], bins, labels=age_groups, include_lowest=True)
kashti['age']
# how this will change the names in dataset based on grouping? (Assigment)
```
