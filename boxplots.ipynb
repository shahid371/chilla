```python
import seaborn
seaborn.set(style="whitegrid")
boat=seaborn.load_dataset("titanic")
seaborn.boxplot(x="class",
                y="fare",
                data= boat)
```

```python
import seaborn
seaborn.set(style="whitegrid")
tip=seaborn.load_dataset("tips")
seaborn.boxplot(x="day",
                y="tip",
                data= tip, saturation=1)
```

```python
import seaborn as sns
sns.set_theme(style="ticks", palette="pastel")
#load data set
tips= sns.load_dataset("tips")
#drawa nested boxplot to show bills by day and time
sns.boxplot(x="day", y="total_bill",
            hue="smoker", palette=["m", "g"],
            data=tips)
sns.despine(offset=10, trim=True)
```

```python

import seaborn as sns
sns.set_theme(style="whitegrid")
#load data set
tips= sns.load_dataset("tips")
#drawa nested boxplot to show bills by day and time
sns.violinplot(x="day", y="total_bill",
            hue="smoker", split=True,
            inner="quart", linewidth=1,
            data=tips)
sns.despine(left=True)
```
