# import libraries

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
phool = sns.load_dataset("iris")
```

```python
sns.lineplot(x="sepal_length", y="sepal_width", data = phool)
#adding title
plt.title("phool_plot")
#adding limit
plt.xlim(4)
plt.ylim(1)
```
